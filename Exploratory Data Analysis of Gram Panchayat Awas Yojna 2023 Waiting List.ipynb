#### Exploring the Pradhan Mantri Awas Yojana 2023.

Welcome to my analysis on the Pradhan Mantri Awas Yojana (PMAY) for 2023! This is the initiative by the Government of India for affordable housing for every citizen. 

#### Why PMAY Matters:
"Envisioning Housing for All by 2022" was the main aim of PMAY. The scheme is not just about constructing buildings but it is about creating spaces where even the economically weaker sections, low-income groups, and middle-income families can thrive. This initiative spans both urban and rural areas, ensuring that no one is left behind.

#### My Mission in Exploratory Data Analysis:
As I dive into the data for 2023, our goal is simple: to understand how well PMAY is reaching the people it aims to serve. 

#### What I am Looking Into:

1. Regional Stories:
   - I am exploring how different states are experiencing the PMAY scheme. 

2. People and Diversity:
   - I am looking in on how different groups, like minorities, scheduled castes, tribes, and others, are experiencing the impact. 

3. The Waiting Game:
   - I have tried to check on communities who are on waiting list. Understanding the backlog helps us spot areas where things might need a little nudge, ensuring that benefits reach those who need them.

4. Minority Impact:
   - How well is PMAY doing in minority communities. I am assessing the extent to which they've embraced the benefits, and I have looked for areas where a bit more outreach and support could make a big difference.

5. Why It Matters:
   - My analysis isn't just numbers and charts. It's a guide for the decision-makers—policymakers, government bodies, and anyone involved in making PMAY shceme work. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the csv file and saving into the df variable.
df = pd.read_csv('Gram Panchayat Awas Yojana 2023.csv', encoding = 'unicode_escape')
```

```python
# Getting the dataframe
df
```

##### Note:
- There is column named 'Unnamed: 13' which is completely empty containing NaN values.
- We don't have any use of that column so we will remove it.

```python
# Removing last column named 'Unnamed: 13' using drop().
df.drop('Unnamed: 13', axis = 1, inplace = True)
```

```python
# Checking whether the column 'Unnamed: 13' is dropped or not.
df.head()
```

##### Note:
- Let's make the column names more readable.

```python
# Changing the Column names:
df.columns = ['State Code','State Name', 'District Code', 'District Name', 'Block Code', 'Block Name', 'Panchayat Code', 'Panchayat Name', 'Minority', 'Others', 'SC', 'ST', 'Total']
df.head(3)
```

```python
# Checking the basic information of the dataframe.
df.info()
```

##### Note:
- Total 13 columns with 270699 entries.
- Data types of columns such as state_code, Others, SC, ST and Total is integer.
- Data types of columns such as state_name, district_name, block_name, panchayat_name and Minority is object.
- Data types of columns such as District_code, Block_Code and panchayat_Code is float.
- Also, columns like District_code, Block_code, Panchayat_code and Panchayat_name are containing null entries.

```python
# Cheecking the data types of the columns of dataframe df.
df.dtypes
```

##### Note:
- Minority column of the dataframe object is of object type.
- If I look at the Minority Column in the df, I can see entries are in whole numbers stating the number of families belonging to minority  class in a region.
- So something has crept into this which is turning this column into object datatype.
- This needs to be cleaned.

```python
# Converting the datatype of column named 'Minority'.
df['Minority'] = pd.to_numeric(df['Minority'], errors = 'coerce')
df.dtypes
```

##### Note:
- I have converted the datatype of this column named 'Minority' from object type to numeric type.
- Whatever thing which crepted this Minority column now has been converted to NaN.
- As NaN is of float datatype, Minority column is returning float datatype.

```python
# Checking for null values in dataframe object.
df.isnull().sum()
```

##### Note:

- Null values in Panchayat Code is 21133.

```python
# Removing the null values.
df.dropna(inplace = True)

# Checking whether the null values are deleted or not.
df.isnull().sum()
```

### EXPLORATORY DATA ANALYSIS

#### STATE NAME

```python
# Setting the figure size
sns.set(rc={'figure.figsize':(18, 7)})

# Plotting the countplot
ax = sns.countplot(x='State Name', data=df, log=True)

# Rotating x-axis labels for better readability
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')

# Setting a proper font size for x-axis labels
ax.tick_params(axis='x', labelsize=12)

# Putting labels on each bar
for i in ax.containers:
    ax.bar_label(i)

# Showing the plot
plt.show()
```

#### MINORITY CLASS

##### Note:
- We can see majority of people belonging from states of Uttar Pradesh, Maharashtra and Madhya Pradesh have yet not received the benefits of Pradhanmantri Gram Awas Yojna scheme.
- Scheme is most effective in states of Andaman & Nicobar, Dadra & Nagar Haveli, Lakshadweep and Puducherry and very less people belonging from this states are in waiting list to avail benefits of the scheme.

```python
# Grouping columns statename and minority class:
grouped_1 = df.groupby('State Name')['Minority'].sum().reset_index()
```

##### Note:
- The above dataframe is grouped state wise for the minority column.

```python
# Plotting the bar graph for the above the above dataframe.
ax = sns.barplot(x = 'State Name',y = 'Minority', data = grouped_1, log = True)

# Rotating x axis labels for better readability.
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45, horizontalalignment = 'right')

# Setting the fontsize of x labels
ax.tick_params( axis = 'x', labelsize = 12)

# putting label on bars
for i in ax.containers:
    ax.bar_label(i, fmt='{:.0f}')
    
# printing the plot
plt.show()
```

##### Note:
- The people belonging from Minority class who have not availed the benefits of Pradhan Mantri Awas Yojna Scheme are highest from state of West Bengal followed by Assam and furter followed by Bihar.
- The people from least minority class who have not avialed benefits are from state of Ladakh and Puducherry followed by Dadar & haveli and further followed by Goa.

##### Conclusion
---------------------------------------------------------------------------------------------------------------------------
1. Uttar Pradesh, Maharashtra, and Madhya Pradesh:
   - In these states, there seems to be a backlog of people who have not yet received the benefits. The respective state governments or local authorities in these states may need to streamline the implementation process, possibly by identifying and addressing bottlenecks in the distribution system.
---------------------------------------------------------------------------------------------------------------------------
2. Andaman & Nicobar, Dadra & Nagar Haveli, Lakshadweep, and Puducherry:
   - These states/union territories appear to be more effective in implementing the scheme. The success in these regions could be attributed to efficient local government bodies, and their strategies could serve as a model for other states.
---------------------------------------------------------------------------------------------------------------------------
3. West Bengal, Assam, Bihar:
   - In West Bengal, Assam, and Bihar, there are a significant number of people from the minority class who have not availed the benefits. The state governments, along with minority affairs departments or similar bodies, may need to work on outreach programs, awareness campaigns, and ensuring accessibility to the scheme.
---------------------------------------------------------------------------------------------------------------------------
4. Ladakh, Puducherry, Dadra & Nagar Haveli, Goa:
   - In Ladakh, Puducherry, Dadra & Nagar Haveli, and Goa, there seems to be a lower number of people from the minority class who have not availed benefits. The local authorities, in collaboration with housing and minority affairs departments, can analyze and share successful strategies with other regions.
---------------------------------------------------------------------------------------------------------------------------
5. Bihar, Maharashtra, Madhya Pradesh:
   - In Bihar, Maharashtra, and Madhya Pradesh, there's a notable number of people from the "Others" class who have not availed benefits. State governments and relevant housing departments may need to review and enhance their outreach efforts and streamline the application process.
---------------------------------------------------------------------------------------------------------------------------
6. West Bengal, Uttar Pradesh, Madhya Pradesh:
   - For people belonging to the Schedule Caste class, West Bengal, Uttar Pradesh, and Madhya Pradesh show higher numbers of those who have not availed the benefits. State governments, along with scheduled caste welfare departments, can focus on targeted campaigns and support mechanisms.
---------------------------------------------------------------------------------------------------------------------------
7. Arunachal Pradesh, Lakshadweep, Puducherry, Telangana:
   - In these states/union territories, the number of people from the Schedule Tribe class not availing benefits is relatively lower. Successful strategies from these regions could be shared with others, and the local tribal affairs departments can collaborate with housing departments for effective implementation.

##### OVERALL
- There is a general trend where West Bengal, Madhya Pradesh, and Bihar consistently have higher numbers of people who have not availed benefits. Addressing the challenges in these states would likely have a significant impact on improving the overall implementation of the scheme.
